#### 6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
def median(n): #define function
    n_sort = sorted(n) #sort values in asc order
    i = (len(n_sort) - 1) // 2 #floor modulus to get int value
    if i % 2 == 0:  #check for even or odd no of values
        return n_sort[i]  #return mid value for odd
    else:
        return (n_sort[i] + n_sort[i + 1]) / 2    
        #return avg of mid values for even

median(df['median_income']) #call function
```

#### 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
def total_bedroom_size(total_bedrooms):
  if total_bedrooms <= 10:
    return "small"
  elif 11 <= total_bedrooms < 1000:
    return "medium"
  else:
    return "large"
```

```python
df1['total_bedroom_size'] = df1["total_bedrooms"].map(total_bedroom_size)
df1
```
