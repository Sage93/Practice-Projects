# Car Price Prediction

### Introduction

In today's fast-paced world, owning a car has become a necessity for many individuals.Car price prediction is a valuable tool for various stakeholders in the automotive industry, including car buyers, sellers, and manufacturers. 

Here are some reasons why car price prediction is needed:

- _Car Buyers_: Car price prediction helps potential buyers make informed decisions about purchasing a vehicle. By knowing the expected price range of a specific car model, buyers can set realistic budgets, negotiate better deals, and avoid overpaying for a vehicle. It allows them to compare prices across different dealerships and make an informed decision.
<br>

- _Car Sellers_: Accurate price prediction enables car sellers, such as dealerships or private sellers, to set competitive prices for their vehicles. It helps them determine the right asking price based on various factors like the car's age, mileage, condition, and market demand. This knowledge improves the chances of selling the car quickly while maximizing profit. 
<br>

- _Insurance Companies_: Car insurance providers can benefit from accurate car price predictions to determine the appropriate insurance premiums for different vehicle models. High-value cars generally have higher premiums, while older or lower-priced vehicles have lower premiums. Having access to reliable price predictions enables insurance companies to estimate the value of a car accurately. 
<br>

- _Financial Institutions_: Lenders and financial institutions utilize car price prediction models to evaluate the value of a vehicle when providing loans or financing options. By accurately predicting a car's price, they can assess the loan-to-value ratio, determine the collateral value, and make informed decisions about loan amounts and interest rates. 
<br>

- _Manufacturers and Dealerships_: Car price prediction is valuable for manufacturers and dealerships when setting initial prices for new car models. It helps them determine competitive pricing strategies based on factors like production costs, market demand, competitor pricing, and projected depreciation rates. Manufacturers and dealerships can also use price prediction models to manage inventory and make data-driven decisions about production and stock allocation. 
<br>

5. _compressionratio:_ The ratio of the volume in the cylinder when the piston is at the bottom of its stroke to the volume when the piston is at the top of its stroke.

There don't seem to be any anamolies in the numerical columns as evident from the above histograms. Almost all of them are skewed in one way or the other which we need to take care of before building models on the dataset.

* Here we can see that, the target variable is right skewed. Log transformation can help in normalization, so this variable can maintain standard scale with other variables:

```python
# Function for log transformation of the column
def log_transform(cars,col):
    for colname in col:
        if (cars[colname] == 1.0).all():
            cars[colname + '_log'] = np.log(cars[colname]+1)
        else:
            cars[colname + '_log'] = np.log(cars[colname])
    cars.info()
```

```python
log_transform(cars,['price'])
```

```python
#Log transformation of the feature 'price'
sns.distplot(cars["price_log"], axlabel="price_log");
```

Now distribution of the car prices looks much more normal.

Before we will start to analyse relationship between variables, we need to take a look at our dependent variable 'price'. We want it to be without outliers and also it's distribution being approximately normal . First, take a look if there any ouliers:

```python
# Boxplot to see outliers

sns.boxplot(x=cars['price_log'])
```

```python
# z-score to check outlier. The data with z-score > 3 or < -3 should be considered as outliers
z_score = stats.zscore(cars['price']).sort_values()
print(z_score)
```

```python
outlier = cars[np.abs(z_score) > 3]
print(outlier)
```

We have 3 observations which we can call outliers. Maybe we would need to remove it, but if we try to understand why these cars are so expensive, we would need to take a look at the rows above and see that there are two buicks and one BMW, they are usually quite exepensive cars, so probably it's not really a 'wrong' data, they are just small amount of really expensive cars, but they shouldn't effect on the accuracy of our model.

Transforming it to normal distribution using power-law which can convert any non-linear distribution to normal using a Box-Cox Transformation

```python
# To show Box-cox Transformation of non-normal data
 
# transform training data & save lambda value
fitted_data, fitted_lambda = stats.boxcox(cars["price_log"])
 
# creating axes to draw plots
fig, ax = plt.subplots(1, 2)
 
# plotting the original data(non-normal) and fitted data (normal)
sns.distplot(cars["price_log"], hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2},
            label = "Non-Normal", color ="green", ax = ax[0])
 
sns.distplot(fitted_data, hist = False, kde = True,
            kde_kws = {'shade': True, 'linewidth': 2},
            label = "Normal", color ="green", ax = ax[1])
 
# adding legends to the subplots
plt.legend(loc = "upper right")
 
# rescaling the subplots
fig.set_figheight(4)
fig.set_figwidth(12)
 
print(f"Lambda value used for Transformation: {fitted_lambda}")
```

```python
sns.boxplot(x=fitted_data)
```

From the above plot, we can say that, data is normally distributed.

### Model Building Using Recursive Feature Elimination (or RFE)

Let's now build a model using recursive feature elimination to select features. We'll first start off with an arbitrary number of features, and then use the statsmodels library to build models using the shortlisted features.

* The Recursive Feature Elimination (or RFE) works by recursively removing attributes and building a model on those attributes that remain.

* It uses the model accuracy to identify which attributes (and combination of attributes) contribute the most to predicting the target attribute.

```python
# RFE with 15 features
lm = LinearRegression()
# rfe_15 = RFE(lm, 15)
rfe_15 = RFE(lm, n_features_to_select=15)

# fit with 15 features
rfe_15.fit(X_train, y_train)

# Printing the boolean results
print(rfe_15.support_)           
print(rfe_15.ranking_)  
```

```python
# making predictions using rfe model
y_pred = rfe_15.predict(X_test)

# r-squared
print(r2_score(y_test, y_pred))
```

```python
# RFE with 13 features
lm = LinearRegression()
rfe_13 = RFE(lm, n_features_to_select=13)

# fit with 15 features
rfe_13.fit(X_train, y_train)

# Printing the boolean results
print(rfe_13.support_)           
print(rfe_13.ranking_)  
```

```python
# making predictions using rfe model
y_pred = rfe_13.predict(X_test)

# r-squared
print(r2_score(y_test, y_pred))
```

The modal has a prediction accuracy of price of about 91%.
