# Data visualization

- Data visualization means drawing graphic displays to show data.
- Data visualization is useful for 
    - data cleaning
    - exploring data structure
    - detecting outliers and unusual groups
    - identifying trends and clusters 
    - spotting local patterns
    - evaluating modeling output
    - and presenting results.

## Count plot
-  Only applicable for categorical data to count observations in each category

```python
sns.countplot(x = 'am', data = mtcars)
# sns.countplot(x='am', data=mtcars) # x axis is am column.
# plt.show()  
```

From the above plot it is clear that 0 has much higher count in comparison to 1

```python








## Adding title and labels for x and y axis
sns.countplot(x='am', data=mtcars) # x axis is am column.
plt.title("Figure 1: Count of automatic and manual car")
plt.xlabel("Automatic or Manual")
plt.ylabel("Total count")
plt.show()
```

This figure tells me that automatic cars are more in count.

```python
ax = sns.countplot(data = mtcars, x ='am').set(title='Figure 1: Count of automatic and manual car', 
                                               xlabel='Automatic vs manual', ylabel='Count')
```

```python
# Marking figures size. It should only be in the begining
plt.figure(figsize=(5,4))

## Adding title and labels for x and y axis
sns.countplot(x='am', data=mtcars, palette = "Set2") # x axis is am column.
plt.title("Count of automatic and manual car")
plt.xlabel("Automatic or Manual")
plt.ylabel("Total count")
plt.ylim(0, 20)
plt.show()

```

```python
# Marking figures size. It should only be in the begining
plt.figure(figsize=(10,5))

## Adding title and labels for x and y axis
sns.countplot(y='am', data=mtcars, palette = "Set3") # y axis is am column.
plt.title("Count of automatic and manual car")
plt.xlabel("Automatic or Manual")
plt.ylabel("Total count")
plt.show()












```

### Comparing 2 categories

```python
sns.set_palette("hls", 1)
sns.countplot(x='am', hue='gear', data=mtcars)
plt.show()
```

## Bar plot
- Statistically it is used to count observations in each category.
- Now we are also using to compare means of two or more population.

```python
# Define a data set
data= {'apples' : 20, 'Mangoes' : 15, 'lemon' : 30, 'oranges' : 10}
names = list(data.keys())
values= list(data.values())

plt.figure(figsize=(8,4))

plt.bar(names,values)
plt.show()
```

```python





# Counting total numbers in each category
count = mtcars['am'].value_counts()
# Checking the numbers
count
# Creating categories based on numbers
AM = ['Automatic', 'Manual']
# Creating plot
fig = plt.figure(figsize=(10,5))
plt.bar(AM, count, color=['cyan', 'pink'])
plt.title("Total number of automatic and manual")
plt.xlabel("Automatic vs Manual")
plt.ylabel("Count")
# show plot
plt.show()
```

```python
# Comparing mean of categorical column
sns.barplot(data = mtcars, x = 'am', y = 'mpg', color = 'cyan',ci = None)
```

## Pie chart
- used to examine the percentage in each category

```python
# Counting total numbers in each category
count = mtcars['am'].value_counts()
# Checking the numbers
count
# Creating categories based on numbers
AM = ['Automatic', 'Manual']
# Creating plot
fig = plt.figure(figsize =(10, 7))
# Show plot
plt.pie(count, labels = AM,autopct='%1.1f%%',shadow=True, startangle=90) # autopct to display the percent value using Python string formatting. For example, autopct='%1.1f%%' means that for each pie wedge, the format string is '1.1f%'
plt.title("Pie Chart")
plt.show()
```

```python
# Counting total numbers in each category
count = mtcars['am'].value_counts()
# Checking the numbers
count
# Creating categories based on numbers
AM = ['Automatic', 'Manual']
# Creating plot
fig = plt.figure(figsize =(10, 7))
# Show plot
plt.pie(count, labels = AM,autopct='%1.1f%%',shadow=True, startangle=90) # autopct to display the percent value using Python string formatting. For example, autopct='%1.1f%%' means that for each pie wedge, the format string is '1.1f%'
plt.title("Pie Chart")
plt.show()
```

## Histogram
- It is used to see the distribution of a numerical value.

```python











plt.hist(mtcars["mpg"], color='#FF2331')
plt.title("Histogram Plot Demo")
plt.xlabel("Range of values")
plt.ylabel("Frequencies")
plt.grid(False)
plt.show()
```

```python
## Adding bins in histogram
plt.hist(mtcars["mpg"], bins=15, color='#FF2331')
plt.title("Histogram Plot Demo")
plt.xlabel("Range of values")
plt.ylabel("Frequencies")
plt.grid(False)
plt.show()
```
